## match statement

The match-case statement in Python is a conditional statement that checks a variable against multiple possible values.
 It works like the switch-case statement in languages like C, C++, or Java.

### Key Points:
1: Introduced in Python 3.10.

2: Performs pattern matching: it matches the variableâ€™s value to a case and executes the corresponding block.

3: A default block can be defined using _ if no case matches.

4: Match statement makes the code cleaner.

5: Match statement uses **case** keyword.

### syntax

#### match variable:


case value1:

  code for value 1

case value2:

  code for value 2

case_:

  default block if no match

### example #1:
(day of the week)

```python
day = "Monday"

match day:
    case "Monday":
        print("Start of the week")
    case "Friday":
        print("Almost weekend")
    case "Sunday":
        print("Rest day")
    case _:
        print("Midweek day")
```

### explanation



*   if day is "Monday" the first block runs.
*   if no case matches, the default block (case _) runs.



### example #2
(number check)

```python
num = 3

match num:
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case _:
        print("Other number")
```

### explanation


*   num is 3, so case three executes.
*   the default case_: will only run if none of the above cases match.



### example #3
food menu

```python
food = "Pizza"

match food:
    case "Burger":
        print("Fast food")
    case "Pizza":
        print("Italian food")
    case "Salad":
        print("Healthy choice")
    case _:
        print("Other food")
```

### explanation


*   match checks the variable's value and runs the block of the matching case.
*   _ work as the default case.



#### points

1: match-case is used mainly for exact value checks.

2: It is cleaner and more readable than if-elif-else when there are many possible values.

3: Use case _: as a default when no case matches.

## 2. Difference between match statement and if

### Match-Case Statement (Python)

1: It is used to match a single variable with multiple specific values or patterns.

2: It works best when checking exact values.

3: It makes code more structured and readable when there are many options.

4: It uses case _: as the default case if no match is found.
It was introduced in Python 3.10, so it does not work in older versions.



### If-Elif-Else Statement (Python)
1: It is used to check logical conditions that return True or False.

2: It can handle complex conditions like greater than, less than, ranges, and logical operators (and, or, not).

3: It is more flexible and suitable for decision-making based on expressions.

4: It uses else: as the default block when no condition is true.
It works in all Python versions.

## nested if_else

A nested if-else statement is a decision-making structure where an if-else statement is placed inside another if or else block.

It is used when a second condition needs to be checked only if the first condition is true (or false).

### nested if else use:
1: When decisions depend on multiple related conditions.

2: When one condition must be checked before another.

3: When solving real-life problems that require step-by-step checking.

#### syntax

if condition1:

    if condition2:

        # block of code

    else:

        # block of code

else:

    # block of code

### example #1
positive, even, odd, negative.

```python
num = 8

if num >= 0:
    if num % 2 == 0:
        print("Positive Even Number")
    else:
        print("Positive Odd Number")
else:
    print("Negative Number")
```

### explanation:
1: First it checks if the number is positive.

2: If true, it then checks whether the number is even or odd.

### example #2
exam result with grade

```python
marks = 85

if marks >= 50:
    if marks >= 80:
        print("Pass with Grade A")
    else:
        print("Pass")
else:
    print("Fail")
```

### explanation
1: First it checks if the student passed.

2: If passed, then it checks if the marks are high enough for Grade A.

### example #3
login system

```python
username = "admin"
password = "1234"

if username == "admin":
    if password == "1234":
        print("Login Successful")
    else:
        print("Incorrect Password")
else:
    print("Username not found")
```

### explanation

1: First it verifies the username.

2: If correct, it checks the password.

### example #4
driving license eligibility

```python
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Eligible for Driving License")
    else:
        print("ID Card Required")
else:
    print("Not Eligible")
```

### Explanation:
1: First age is checked.

2: If age is valid, then ID availability is checked.

### example #5
online shoping discount

```python
amount = 7000
member = True

if amount > 5000:
    if member:
        print("You get 20% discount")
    else:
        print("You get 10% discount")
else:
    print("No discount available")
```

### Explanation:
1: First purchase amount is checked.

2: If amount is enough, then membership is checked for extra discount.
