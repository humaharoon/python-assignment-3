# Loops


## for loop:
Afor loop is used to repeat a block of code a specific number of times or to go through item in a sequence like a list, tuple or string.

### syntax:
for variable in sequence:

    # code block

### forexample:




```python
for i in range(1,5):
    print(i)
```

## while loop:
A while loop runs as long as a condition is true. It checks the condition first. If it is true the code runs. If it is false the loop stops.

### syntax:
while condition:

    # code block
### forexample:


```python
x=1
while x<=5:
  print(x)
  x+=1
```

## break statement:
break is used to stop the loop immediately even if the condition is still true.

### forexample:

```python
for i in range(1,5):
    if i==3:
      break
    print(i)
```

## continue statement:
continue statement is used to skip the current iteration and move to the next one.

### forexample:


```python
for i in range(1,5):
  if i==3:
    continue
  print(i)
```

1: **for loop** is used when number of repetitions is known.

2: **while loop** is used when repetition depends on condition.

3: **break** is used to stops the loop completely.

4: **continue** is used to skip on iteration and continues the loop.

### forexample:


```python
string="aabbcc"
count=3
while True:
  if string[0]=='a':
     string=string[2:]
  elif string[-1]=='b':
     string=string[:2]
  else:
    count+=1
    break
print(string)
print(count)
```
